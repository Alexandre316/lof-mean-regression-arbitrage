#### 2.计算年化折价率

```python
discount_ratio = pd.read_csv("./Data/LOF溢价率.CSV",index_col = 0)
discount_ratio.index = pd.to_datetime(discount_ratio.index)
```

#### 3.观测折价变动

```python
for lof in discount_ratio.columns:
    print(f"processing {lof}")
    lof_discount_ratio = discount_ratio.loc[:,lof].dropna()
    if lof_discount_ratio.empty:
        continue
    lof_discount_ratio_mean = np.nanmean(lof_discount_ratio)
    plt.figure(figsize=(20,5),dpi = 200)
    plt.plot(lof_discount_ratio,"-",color = "#0000FF",linewidth = 0.5,label = "Net Value")
    plt.hlines(lof_discount_ratio_mean,lof_discount_ratio.index[0],lof_discount_ratio.index[-1],linestyles = "solid",color = "red",linewidth = 1,label = "Mean")
    plt.legend([f"{lof} Discount Ratio"], loc = 'upper right',fontsize = 12.5,shadow = "gray",fancybox = False)
    plt.savefig(f"./Plots/Discount/{lof[:6]} Discount Ratio.png")
    plt.close()
#     plt.show()
```
