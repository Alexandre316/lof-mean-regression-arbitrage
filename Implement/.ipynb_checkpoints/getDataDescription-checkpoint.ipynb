#### 2.封闭开放数据 

```python
# 开放和赎回状态转化为具体的日期
redemption_status = pd.read_excel("./Data/RedemptionStatus.xlsx",index_col = 0)
subscribe_status = pd.read_excel("./Data/SubscribeStatus.xlsx",index_col = 0)
MaturitySuppliment = pd.read_excel('./Data/MaturitySuppliment.xlsx')
```

```python
MaturitySuppliment.F_INFO_WINDCODE = MaturitySuppliment.F_INFO_WINDCODE.astype(str)+".OF"
MaturitySuppliment.index = MaturitySuppliment.F_INFO_WINDCODE
```

```python
redemption_status.drop("F_INFO_WINDNAME",axis = 1,inplace = True)
subscribe_status.drop("F_INFO_WINDNAME",axis = 1,inplace = True)
```

```python
# 赎回状态
all_redemption_status = list()
for i in redemption_status.index:
    all_redemption_status.extend(redemption_status.loc[i].unique())
all_redemption_status = list(set(all_redemption_status))

# 申购状态
all_subscribe_status = list()
for i in subscribe_status.index:
    all_subscribe_status.extend(subscribe_status.loc[i].unique())
all_subscribe_status = list(set(all_subscribe_status))
```

```python
# 赎回状态 '封闭期', '开放赎回', '未成立', '暂停赎回'
# 申购状态 '未成立', '暂停大额申购', '暂停申购', '封闭期', '开放申购'(不重要，只要保证是已经成立的基金即可)
redemption_status.replace({"封闭期":2,"未成立":3,"暂停赎回":2,"开放赎回":0},inplace = True)
subscribe_status.replace({'未成立':0, '暂停大额申购':1, '暂停申购':1, '封闭期':1, '开放申购':1},inplace = True)
```

```python
redemption_status_change = redemption_status.diff(axis = 1)
redemption_status_change.dropna(inplace = True,axis = 1)
```

```python
# 记录每日距离最近的
mature_period = pd.DataFrame(index = redemption_status.index,columns = redemption_status.columns)
```

```python
# 开放转封闭求封闭期
for i in redemption_status_change.index:
    end_date = None
    
    print(f"processing {i}")

    for begin_date in redemption_status_change.loc[i,redemption_status_change.loc[i] == 2].index:
        for end_date_ in redemption_status_change.loc[i,redemption_status_change.loc[i] == -2].index:
            if end_date_ > begin_date:
                end_date = end_date_
                break
            try:    
                end_date = MaturitySuppliment.loc[i,"MATURITY_DATE"]
            except KeyError:
                end_date = pd.to_datetime("2022-03-31")
            
#         print(f"{i}",begin_date,end_date,(end_date-begin_date)/np.timedelta64(1,'D')) # 每次暂停赎回和开放赎回都是可预见的，每次需要的是[0.5,1]的基金，那种短期的完全不考虑
        
        # 到期日存储
        for j in redemption_status_change.loc[i,begin_date:end_date].index:
            mature_period.loc[i,j] = (end_date - j) / np.timedelta64(1,'D')
    
        end_date = None
```

```python
# 未成立转封闭，且封闭期未结束
for i in redemption_status_change.index:
    
    end_date = None
    begin_date = None
    
    if len(redemption_status_change.loc[i,redemption_status_change.loc[i] != 0]) == 1: # 如果考核期内只有一次状态改变
        if not redemption_status_change.loc[i,redemption_status_change.loc[i] == -1].empty: # 且状态改变为未成立转封闭且封闭期未结束
            
            begin_date = redemption_status_change.loc[i,redemption_status_change.loc[i] == -1].index[0]
            try:    
                end_date = MaturitySuppliment.loc[i,"MATURITY_DATE"]
            except KeyError:
                end_date = pd.to_datetime("2022-03-31")
            
            print(f"{i}",begin_date,end_date,(end_date-begin_date)/np.timedelta64(1,'D'))
            
            for j in redemption_status_change.loc[i,begin_date:end_date].index:
                mature_period.loc[i,j] = (end_date - j) / np.timedelta64(1,'D')
```

```python
# 封闭转开放，且维持开放期不变
for i in redemption_status_change.index:
    
    end_date = None
    begin_date = None
    
    if len(redemption_status_change.loc[i,redemption_status_change.loc[i] != 0]) == 1: # 如果考核期内只有一次状态改变
        if not redemption_status_change.loc[i,redemption_status_change.loc[i] == -2].empty: # 且状态改变为未成立转封闭且封闭期未结束
            
            begin_date = pd.to_datetime("2013-01-01")
            end_date = redemption_status_change.loc[i,redemption_status_change.loc[i] == -2].index[0]
            
            print(f"{i}",begin_date,end_date,(end_date-begin_date)/np.timedelta64(1,'D'))
            
            for j in redemption_status_change.loc[i,begin_date:end_date].index:
                mature_period.loc[i,j] = (end_date - j) / np.timedelta64(1,'D')
```

```python
mature_period.to_excel("./Output/MaturePeriod.xlsx")
```
