#### 1.数据导入

```python
# 净值、价格和成交量数据
lof_open = pd.read_csv("./Data/LOF开盘价.CSV",index_col = 0)
lof_close = pd.read_csv("./Data/LOF收盘价.CSV",index_col = 0)
lof_high = pd.read_csv("./Data/LOF最高价.CSV",index_col = 0)
lof_low = pd.read_csv("./Data/LOF最低价.CSV",index_col = 0)
lof_nav = pd.read_csv("./Data/LOF复权净值.CSV",index_col = 0)
lof_volume = pd.read_csv("./Data/LOF成交量.CSV",index_col = 0)

# 将SZ,SH结尾的场内LOF转换成OF结尾的
lof_open.columns = lof_open.columns.map(lambda x:x[:6]+".OF")
lof_close.columns = lof_close.columns.map(lambda x:x[:6]+".OF")
lof_high.columns = lof_high.columns.map(lambda x:x[:6]+".OF")
lof_low.columns = lof_low.columns.map(lambda x:x[:6]+".OF")
lof_nav.columns = lof_nav.columns.map(lambda x:x[:6]+".OF")
lof_volume.columns = lof_volume.columns.map(lambda x:x[:6]+".OF")

lof_open.index = pd.to_datetime(lof_open.index)
lof_close.index = pd.to_datetime(lof_close.index)
lof_high.index = pd.to_datetime(lof_high.index)
lof_low.index = pd.to_datetime(lof_low.index)
lof_nav.index = pd.to_datetime(lof_nav.index)
lof_volume.index = pd.to_datetime(lof_volume.index)
```

```python
# 导入到期日数据
maturity_period = pd.read_excel("./Output/MaturePeriod.xlsx",index_col = 0)

# 导入申购状态数据
redemption_status = pd.read_excel("./Data/RedemptionStatus.xlsx",index_col = 0)
subscribe_status = pd.read_excel("./Data/SubscribeStatus.xlsx",index_col = 0)

# 导入beta_norm指标
signals = pd.read_excel("./Output/BetaRightInfo.xlsx",index_col = 0)
```

#### 2.计算年化折价率

```python
discount_ratio = lof_close/lof_nav-1

annual_discount_ratio = discount_ratio*365/np.transpose(maturity_period)

annual_discount_ratio.dropna(how = "all",axis = 1,inplace = True)
```

#### 3.构建买卖信号

```python
start_date = pd.to_datetime("2014-01-01")
end_date = pd.to_datetime("2022-03-25")


'''parameters'''
upper_year = 365 # 距离下次开放最远的购买时间
lower_year = 180 # 距离下次开放最近的购买时间

single_position_limit = 0.1 # 个基最大的仓位限制
single_buy_limit = 1/30 # 每次购买3.33%成的仓位

profit_stop_upper = 0.15 # 个基止盈
profit_stop_lower = 0.05  # 个基止损

S = 0.7 # 买入指标

adding_percent = 0.97 # 当购买价格小于或等于该值倍数时才能加仓平摊成本
target_discount = -0.05 # 买入目标年化折价率
```

```python
# 记录仓位变化
position_holdings = pd.DataFrame(index = discount_ratio.index,columns = discount_ratio.columns,data = 0.0)
buy_price = pd.DataFrame(index = discount_ratio.index,columns = discount_ratio.columns)

'''variables'''
capital = 100000 # 本金
```

```python
for date in annual_discount_ratio.loc[start_date:end_date].index: # 获取每日折价率
    
    # 对年化折价率、到期时间和择时指标做出限制
    daily_discount_lof = annual_discount_ratio.loc[date,annual_discount_ratio.loc[date,:]<target_discount].dropna()
    in_maturity_lof = maturity_period.loc[(maturity_period.loc[:,date]>lower_year)&(maturity_period.loc[:,date]<=upper_year),date]
    timing_signal = signals.loc[signals.loc[:,date]>S,date]
    
    # 获取满足所有指标条件在池子中的lof集合，之后判断购买条件
    in_pool_lof = set(daily_discount_lof.index) & set(in_maturity_lof.index) & set(timing_signal.index)
    # 先卖出，后购买
    # 卖出、止损操作，卖就以当天的收盘价卖出，不再延后
    in_position_lof = position_holdings.loc[date,position_holdings.loc[date] != 0] # 获取当前仓位
    for tmp_position_lof in in_position_lof.index:
        
        tmp_sell_price = lof_close.loc[date,tmp_position_lof] # 获取当前仓位每支LOF的价格
        tmp_discount = discount_ratio.loc[date,tmp_position_lof] # 当前LOF的折价率，不是年化折价率
        
        if maturity_period.loc[tmp_position_lof,date] == 0: # 假如到期，直接卖出
            print(f"sell {tmp_position_lof}|sell date:{date}|sell price:{tmp_sell_price}")
            position_holdings.loc[date:,tmp_position_lof] = 0
            
        # 假如触发了止损条件
        # 为了简化，假设每次购买都是固定资金规模，那么成本价就可以平均，如果没有成本价就不用止盈止损
        tmp_cost = np.nanmean(buy_price.loc[:date,tmp_position_lof]) # 过往的平均成本
        # 止盈条件
        if tmp_sell_price >= tmp_cost*(1+profit_stop_upper): # 如果盈利达到15%
            print(f"sell {tmp_position_lof}|sell date:{date}|sell price:{tmp_sell_price}|(止盈)")
            position_holdings.loc[date:,tmp_position_lof] = 0
        # 止损条件
        if tmp_sell_price/tmp_cost-1 < tmp_discount: # 如果亏损大于折价率
            print(f"sell {tmp_position_lof}|sell date:{date}|sell price:{tmp_sell_price}|(止损)")
            position_holdings.loc[date:,tmp_position_lof] = 0
    
    # 判断价格条件：下次购买的价格必须要低于本次购买价格的3%，每次购买5%仓位
    if not in_pool_lof:
        continue
    else:
        for tmp_lof in list(in_pool_lof):
            for after in range(1,10): # 只在未来10天挂单
                tmp_buy_price = None
                tmp_volume = lof_volume.loc[date:,tmp_lof].iloc[after]
                if pd.isnull(tmp_volume): # 如果次日么有成交，再往后挂
                    continue
                else:
                    tmp_buy_price = lof_high.loc[date:,tmp_lof].iloc[after] # 获取决定报单之后第一个非0成交量当日的最高价(最差的价格)
                    tmp_buy_date = lof_high.loc[date:,[tmp_lof]].iloc[after].name # 获取本次购买的时间，可能会延后几天
                    tmp_single_position_count = position_holdings.loc[tmp_buy_date,tmp_lof] # 统计某只基金头上的总仓位
                    
                    try:
                        last_buy_price = buy_price.loc[:tmp_buy_date,tmp_lof].dropna()[-1] # 获取上次购买价
                    except IndexError:
                        last_buy_price = 0
                        
                    # 若存在非0的上次购买价，且仓位大于0.1成，或者最新购买价尚未够低，或者仓位满了，就不再买了
                    if (last_buy_price) and (tmp_single_position_count >= single_position_limit or \
                                             tmp_buy_price > last_buy_price*adding_percent or in_position_lof.sum() >= 1):
                        break 
                    
                    print(f"buy {tmp_lof}|signal date:{date}|buy date:{tmp_buy_date}|buy price:{tmp_buy_price}")
                    
                    # 记录仓位和购买价格
                    position_holdings.loc[tmp_buy_date:,tmp_lof] += single_buy_limit
                    buy_price.loc[tmp_buy_date,tmp_lof] = tmp_buy_price
                    break
#             print(position_holdings.loc[tmp_buy_date].sum()) # 大概1~3成仓位运作
```

#### 4.统计净值变动

```python
lof_returns = lof_close.pct_change()
```

```python
daily_returns = pd.Series((lof_returns * (position_holdings.asfreq('B').ffill().shift(1))).sum(1),\
index = pd.to_datetime(lof_returns.index), name = 'Returns').fillna(0)
```

```python
fig = plt.figure(figsize = (18,6),dpi = 150)
ax1 = fig.add_subplot(111)
ax1.plot(np.cumprod((1 + daily_returns).loc[start_date:]), "-",color = "#0000FF",linewidth = 1,label = "Net Value")#非预测对比序列
plt.xticks(fontproperties='Times New Roman', size = 11.5)
plt.yticks(fontproperties='Times New Roman', size = 11.5)
ax1.set_ylabel("Net Value (Adjusted)",fontsize = 12,weight = 'bold')
ax1.set_xlabel("Years",fontsize = 12,weight = 'bold')
ax1.legend(["Net Value"], loc = 'upper left',fontsize = 12.5,shadow = "gray",fancybox = False)

ax2 = ax1.twinx()#真实指数序列
plt.ylim(-0.15, 1)
ax2.plot((position_holdings.sum(axis = 1).loc[start_date:]), "-",color = "orange",linewidth = 1,label = "Position Percentage")#非预测对比序列
ax2.set_ylabel("Position",fontsize = 12,weight = 'bold')

plt.legend(loc = 'lower right',fontsize = 12.5,shadow = "gray",fancybox = False)
```
