#### 2.回测数据生成模块

```python
tmp_fund = "501203.OF"

N = 16
M = 200

start_date = pd.to_datetime("2014-01-01")
end_date = pd.to_datetime("2022-03-29")

beta_right_threshold = 0.7
```

```python
# 每一天评估每只基金，记录每支基金的beta_right
beta_right_info = pd.DataFrame(index = redemption_status.index,columns = redemption_status.loc[:,start_date:end_date].columns)
```

```python
for i in beta_right_info.columns:
    for tmp_fund in beta_right_info.index:
        # 判断是否为封闭基金，只考虑封基折价方法
        if pd.isnull(maturity_period.loc[tmp_fund,i]):
            continue
        # 每个基金都要计算一次回归       
        data_info = pd.DataFrame()
        beta_norm = 0
        RSRS_R2 = 0
        beta_right = 0
        # 获取前150个连续交易日得开收低高数据，有交易日的数据日长度小于M，则认定数据不全难以进一步分析而跳过
        # 因为去除无成交的交易日，可能数据不连续
        data_info.loc[:,"volume"] = lof_volume.loc[:i,tmp_fund].dropna().iloc[-(M+N):] # 去除没有成交的，截止至当日的共M+N个交易日的数据，需要计算M个beta
        try:
            deal_start_date = data_info.index[0] # 如果数据是空则不用考虑
        except IndexError:
            continue
        data_info.loc[:,"open"] = lof_open.loc[:i,tmp_fund].loc[deal_start_date:] # 之后再从有交易的交易日中获取数据
        data_info.loc[:,"close"] = lof_close.loc[:i,tmp_fund].loc[deal_start_date:]
        data_info.loc[:,"high"] = lof_high.loc[:i,tmp_fund].loc[deal_start_date:]
        data_info.loc[:,"low"] = lof_low.loc[:i,tmp_fund].loc[deal_start_date:]
        data_info.loc[:,"nav"] = lof_nav.loc[:i,tmp_fund].loc[deal_start_date:]
        # 去除成交为空的交易日
        data_info.dropna(how = "any",inplace = True) # 需要解决没有成交的数据
        # 重设index
        data_info['date'] = data_info.index
        data_info.reset_index(inplace = True,drop = True)
        # 去除不满M日的数据条目
        if len(data_info) < M+N:
            continue
        # 指标计算
        for j in range(N,len(data_info)):
            slope_window = data_info.loc[j-N:j-1,:]
            model = sml.ols(formula = "high~low",data = slope_window)
            result = model.fit()
            # 存储序列指标
            data_info.loc[j-1,'beta'] = result.params[1]
            data_info.loc[j-1,'R2'] = result.rsquared
        # 标准分
        beta_norm = (data_info.beta.iloc[-2] - data_info.loc[N-1:N+M-2,'beta'].mean())/(data_info.loc[N-1:N+M-2,'beta'].std())
        RSRS_R2 = beta_norm*data_info.R2.iloc[-2]
        beta_right = RSRS_R2*data_info.beta.iloc[-2]
        
        print(f"{i},{tmp_fund},{beta_right}") # 数据存储
        beta_right_info.loc[tmp_fund,i] = beta_right # shift(1)添加在前面
```

```python
beta_right_info.to_excel("./Output/tput/BetaRightInfo.xlsx")
```

```python
data_info = pd.DataFrame()
```

```python
data_info.loc[:,"open"] = lof_open.loc[:,tmp_fund]
data_info.loc[:,"close"] = lof_close.loc[:,tmp_fund]
data_info.loc[:,"high"] = lof_high.loc[:,tmp_fund]
data_info.loc[:,"low"] = lof_low.loc[:,tmp_fund]
data_info.loc[:,"nav"] = lof_nav.loc[:,tmp_fund]
data_info.loc[:,"volume"] = lof_volume.loc[:,tmp_fund]
```

```python
data_info.dropna(inplace = True)

data_info['date'] = data_info.index
data_info.reset_index(inplace = True,drop = True)
```

```python
data_info["beta"] = 0
data_info["R2"] = 0
```

```python
for i in range(N,len(data_info)):
    
    slope_window = data_info.loc[i-N:i,:]
    model = sml.ols(formula = "high~low",data = slope_window)
    result = model.fit()
    
    data_info.loc[i+1,'beta'] = result.params[1]
    data_info.loc[i+1,'R2'] = result.rsquared
    
# 日收益率
data_info['ret'] = data_info.close.pct_change(1)

# 标准分
data_info['beta_norm'] = (data_info.beta - data_info.beta.rolling(M).mean().shift(1))/(data_info.beta.rolling(M).std().shift(1))
# 最初300日的标准分，可能不构成买点
for i in range(M):
    data_info.loc[i,'beta_norm'] = (data_info.loc[i,'beta'] - data_info.loc[:i-1,'beta'].mean())/data_info.loc[:i-1,'beta'].std()
data_info.loc[N+1,'beta_norm'] = 0 # 去除inf

data_info['RSRS_R2'] = data_info.beta_norm*data_info.R2

data_info.fillna(0,inplace = True)

data_info['beta_right'] = data_info.RSRS_R2*data_info.beta
```
