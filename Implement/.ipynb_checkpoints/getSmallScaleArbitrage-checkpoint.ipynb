#### 2.计算价格偏移

```python
start_date = pd.to_datetime("2013-12-31")
end_date = pd.to_datetime("2022-03-25")
```

```python
discount_ratio = pd.read_csv("./Data/LOF溢价率.CSV",index_col = 0)
discount_ratio.index = pd.to_datetime(discount_ratio.index)
discount_ratio.columns = discount_ratio.columns.map(lambda x:x[:6]+".OF")
```

```python
bias_info = pd.DataFrame(index = discount_ratio.index,columns = discount_ratio.columns,data = False)
```

```python
for date in trade_date.loc[(trade_date.DATE>=start_date) & (trade_date.DATE<=end_date)].DATE:
    for lof in discount_ratio.columns:
        tmp_discount_ratio = discount_ratio.at[date,lof]
        if pd.isnull(tmp_discount_ratio):
            continue
        period_discount_ratio = discount_ratio.loc[:date,lof].dropna()
        if len(period_discount_ratio) < 100:
            continue
        std = np.std(period_discount_ratio)
        mean = np.nanmean(period_discount_ratio)
        if tmp_discount_ratio < 0 and tmp_discount_ratio < mean - 1*std:
            print(f"{date},{lof},{mean},{std},{tmp_discount_ratio}")
            bias_info.at[date,lof] = True
```

#### 3.构建买卖信号

```python
@dataclass
class FundObject:
    vt_symbol:str = ""
    buy_price:float = 0.0
    volume:float = 0.0
    
    def __post_init__(self):
        self.price:float = self.buy_price
        self.last_price:float = self.price
        self.pnl:float = (self.price - self.buy_price)*self.volume
        self.daily_pnl:float = (self.price - self.last_price)*self.volume
        self.status:str = "HOLDING" # HOLDING, SOLD, ASK, LOCKING
        self.sequence:int = 1
        self.buy_discount:float = 0.0
        self.last_discount_avg:float = 0.0
        self.latest_nav:float = 0.0
```

```python
# 记录仓位变化
'''account'''
capital:float = 10000.0 # 本金
available:float = capital
in_position_lof:dict = {}
backtest_result:pd.DataFrame = pd.DataFrame(columns = ["ACCOUNT_BALANCE"])
'''variables'''

'''paramaters'''
upper_year:int = 365 # 距离下次开放最远的购买时间
lower_year:int = 90 # 距离下次开放最近的购买时间

single_position_limit:float = 0.1 # 个基最大的仓位限制
single_buy_limit:float = 0.01 # 每次购买2%成的仓位

profit_stop_upper:float = 0.15 # 个基止盈
profit_stop_lower:float = 0.05  # 个基止损

buy_cost:float = 0.003
sell_cost:float = 0.003
```

```python
for date in trade_date.loc[(trade_date.DATE>=start_date) & (trade_date.DATE<=end_date)].DATE:
    """"""    
    # to unlock
    for to_unlock_lof in in_position_lof.keys():
        if in_position_lof[to_unlock_lof].status == "LOCKING":
            in_position_lof[to_unlock_lof].status = "HOLDING"
            
    for to_bid_lof in list(in_position_lof.keys()):
        if in_position_lof[to_bid_lof].status == "BID": 
            # sell at open price 
            sell_price = lof_open.loc[date,to_bid_lof]
            sell_volume = 100
            if pd.isnull(sell_price): # no volume the whole day, bid in the next day
                continue
            available += sell_price*sell_volume*(1-sell_cost)
            capital = capital - sell_price*sell_volume*sell_cost + (sell_price-in_position_lof[to_bid_lof].price)*in_position_lof[to_bid_lof].volume
            del in_position_lof[to_bid_lof]
            print(f"arbitrage {to_bid_lof}|sell date:{date}|sell price:{sell_price}")
        
    # buy at open close, at the beginning of the day
    for to_buy_lof in list(in_position_lof.keys()):
        if in_position_lof[to_buy_lof].status == "ASK": # lof to buy
            buy_price = lof_open.loc[date,to_buy_lof]
            buy_volume = 100 # fix volume (free from market volume risks)
            if pd.isnull(buy_price) or buy_price*buy_volume*(1+buy_cost)>available or \
            buy_price/in_position_lof[to_buy_lof].latest_nav-1>(in_position_lof[to_buy_lof].last_discount_avg-(buy_cost+sell_cost+0.001)): # opportuniry disappears at once 
                del in_position_lof[to_buy_lof] # miss oppotunity or not enough money
                continue
            # deal
            in_position_lof[to_buy_lof].price,in_position_lof[to_buy_lof].buy_price,in_position_lof[to_buy_lof].last_price = buy_price,buy_price,buy_price
            in_position_lof[to_buy_lof].volume = buy_volume
            print(f"buy {to_buy_lof}|buy date:{date}|buy price:{buy_price}|buy volume:{buy_volume}")
            # account data
            available -= in_position_lof[to_buy_lof].volume*in_position_lof[to_buy_lof].price*(1+buy_cost)
            in_position_lof[to_buy_lof].pnl -= in_position_lof[to_buy_lof].volume*in_position_lof[to_buy_lof].price*buy_cost
            in_position_lof[to_buy_lof].daily_pnl -= in_position_lof[to_buy_lof].volume*in_position_lof[to_buy_lof].price*buy_cost
            in_position_lof[to_buy_lof].status = "LOCKING"
    
    # sell, in the middle of the day
    for to_sell_lof in in_position_lof.keys():
        if in_position_lof[to_sell_lof].status == "HOLDING" and not pd.isnull(lof_low.at[date,to_sell_lof]):
            # profit stop
            if lof_low.at[date,to_sell_lof]/in_position_lof[to_sell_lof].buy_price-1 >= profit_stop_upper:
                print(f"profit {to_sell_lof}|sell date:{date}|sell price:{lof_low.at[date,to_sell_lof]}")
                in_position_lof[to_sell_lof].status = "SOLD"
            
            # loss stop
            elif lof_low.at[date,to_sell_lof]/in_position_lof[to_sell_lof].buy_price-1 <= -profit_stop_lower:
                print(f"loss {to_sell_lof}|sell date:{date}|sell price:{lof_low.at[date,to_sell_lof]}")
                in_position_lof[to_sell_lof].status = "SOLD"
            
            # pnl compute
            if in_position_lof[to_sell_lof].status == "SOLD":
                price = lof_low.at[date,to_sell_lof]
                in_position_lof[to_sell_lof].last_price = in_position_lof[to_sell_lof].price
                in_position_lof[to_sell_lof].price = price
                in_position_lof[to_sell_lof].daily_pnl = (in_position_lof[to_sell_lof].price - \
                                                          in_position_lof[to_sell_lof].last_price)*in_position_lof[to_sell_lof].volume
                in_position_lof[to_sell_lof].pnl += in_position_lof[to_sell_lof].daily_pnl
    
    # adjust, at the end of the day
    for to_adjust_lof in in_position_lof.keys():
        # daily pnl compute
        if in_position_lof[to_adjust_lof].status in ["HOLDING","LOCKING"]:
            price = filled_lof_close.at[date,to_adjust_lof]
            in_position_lof[to_adjust_lof].last_price = in_position_lof[to_adjust_lof].price
            in_position_lof[to_adjust_lof].price = price
            in_position_lof[to_adjust_lof].daily_pnl = (in_position_lof[to_adjust_lof].price - \
                                                        in_position_lof[to_adjust_lof].last_price)*in_position_lof[to_adjust_lof].volume
            in_position_lof[to_adjust_lof].pnl += in_position_lof[to_adjust_lof].daily_pnl
            in_position_lof[to_adjust_lof].latest_nav = price/(1+discount_ratio.at[date,to_adjust_lof])
            in_position_lof[to_adjust_lof].last_discount_avg = np.nanmean(discount_ratio.loc[:date,to_adjust_lof].dropna())
            
        # discount ratio converge
        lof_discount_ratio_mean = np.nanmean(discount_ratio.loc[:date,to_adjust_lof].dropna())
        if discount_ratio.at[date,to_adjust_lof] >= lof_discount_ratio_mean and \
        in_position_lof[to_adjust_lof].price/in_position_lof[to_adjust_lof].buy_price-1>(sell_cost+buy_cost+0.001) and \
        in_position_lof[to_adjust_lof].status == "HOLDING": # arbitrage success
            in_position_lof[to_adjust_lof].status = "BID"
    
    # settle order and cancel
    for to_settle_lof in list(in_position_lof.keys()):
        # account daily pnl
        if in_position_lof[to_settle_lof].status in ["LOCKING","SOLD","HOLDING","BID"]:
            capital += in_position_lof[to_settle_lof].daily_pnl
        
        # cancel sold records
        if in_position_lof[to_settle_lof].status == "SOLD":
            available += in_position_lof[to_settle_lof].price*in_position_lof[to_settle_lof].volume*(1-sell_cost)
            capital -= in_position_lof[to_settle_lof].price*in_position_lof[to_settle_lof].volume*sell_cost
            del in_position_lof[to_settle_lof]
            
    # buy condition, prepare to ask the next day
    in_pool_lof = bias_info.loc[date,bias_info.loc[date]].index.tolist()
    for to_ask_lof in in_pool_lof:
        if to_ask_lof not in in_position_lof.keys():
            in_position_lof[to_ask_lof] = FundObject(vt_symbol = to_ask_lof)
            in_position_lof[to_ask_lof].status = "ASK"
            in_position_lof[to_ask_lof].buy_discount = discount_ratio.at[date,to_ask_lof]
            in_position_lof[to_ask_lof].latest_nav = filled_lof_close.at[date,to_ask_lof]/(1+in_position_lof[to_ask_lof].buy_discount)
            in_position_lof[to_ask_lof].last_discount_avg = np.nanmean(discount_ratio.loc[:date,to_ask_lof].dropna())
    
    backtest_result.loc[date,"ACCOUNT_BALANCE"] = capital
```

#### 4.统计净值变动

```python
fig = plt.figure(figsize = (18,6),dpi = 150)
ax1 = fig.add_subplot(111)
ax1.plot(backtest_result.ACCOUNT_BALANCE, "-",color = "#0000FF",linewidth = 1,label = "Net Value")#非预测对比序列
plt.xticks(fontproperties='Times New Roman', size = 11.5)
plt.yticks(fontproperties='Times New Roman', size = 11.5)
ax1.set_ylabel("Net Value (Adjusted)",fontsize = 12,weight = 'bold')
ax1.set_xlabel("Years",fontsize = 12,weight = 'bold')
ax1.legend(["Net Value"], loc = 'upper left',fontsize = 12.5,shadow = "gray",fancybox = False)

plt.legend(loc = 'lower right',fontsize = 12.5,shadow = "gray",fancybox = False)
# plt.savefig('./Plots/Arbitrage(SmallScale)',bbox_inches = 'tight')
```
